### This Dataset is highly unbalanced

    1 ---> Fraudulent Transaction

```python
#Separating the data for analysis
legit = ccdf_dt[ccdf_dt.Class == 0]
fraud = ccdf_dt[ccdf_dt.Class == 1]
print(legit.shape)
print(fraud.shape)
```

### Under-Sampling
#### Building a sample dataset containing similar distribution of normal transcations and fraudulent transactions

```python
legit_samp = legit.sample(n=492)
```

```python
#Concatenation two dataFrames
new_dataset = pd.concat([legit_samp, fraud], axis=0)
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

#### Splitting the data into Features and Targets

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

#### Splitting the data into Training and Test data

```python
X_train, x_test, Y_train, y_test = train_test_split(X, Y, stratify=Y, random_state=2, test_size=0.2)
```

### Model Training
#### Logistic Regression

```python
model = LogisticRegression()
```

```python
# Training the Logistic Regression modelto the training Data
model.fit(X_train, Y_train)
```

### Model Evaluation
#### Accuracy Score

```python
#Accuracy on training data
x_train_pred =  model.predict(X_train)
training_dt_acc = accuracy_score(x_train_pred, Y_train)

print('Accuracy on Training Data :', training_dt_acc)
```

```python
#Accuracy on test data
x_test_pred =  model.predict(x_test)
test_dt_acc = accuracy_score(x_test_pred, y_test)

print('Accuracy on Test Data :', test_dt_acc)
```
