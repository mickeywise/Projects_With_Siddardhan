```python
#Importing library dependencies
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
#loading the data set
diabetes_df = pd.read_csv('C:/Users/MACHIRA/Projects_With_Siddardham/data/diabetes.csv')
diabetes_df.head()
```

    0 ----> Non-Diabetic
    1 ----> Diabetic

```python
diabetes_df.groupby('Outcome').mean()
```

```python
#separating the features and the target variable
X = diabetes_df.drop(columns='Outcome', axis=1)
Y = diabetes_df['Outcome']
```

## Data Standardization

```python
scaler = StandardScaler()
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
standardized_data = scaler.fit_transform(X)
standardized_data
```

```python
#You can fit and transform in one line
#standardized_data = scaler.fit_transform(X)
#standardized_data#
```

```python
X = standardized_data
X
```

### Train Test Split

```python
X_train, x_test, Y_train, y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
print(X.shape, X_train.shape, x_test.shape)
```

### Training the model

```python
classifier = svm.SVC(kernel='linear')
classifier.fit(X_train, Y_train)
```

### Model Evaluation
#### Accuracy Score

```python
#accuracy score on the training data
X_train_pred = classifier.predict(X_train)
training_acc = accuracy_score(X_train_pred, Y_train)
print('Accuracy score of the training data :', training_acc)
```

```python
#accuracy score on the test data
X_test_pred = classifier.predict(x_test)
test_acc = accuracy_score(X_test_pred, y_test)
print('Accuracy score of the test data :', test_acc)
```

### Making a Predictive System

```python
input_dta = (3,158,76,36,245,31.6,0.851,28)
#let's change the input data into numpy array
input_data_as_numpyarray = np.asarray(input_dta)
#reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpyarray.reshape(1, -1) 

#standardize the input data
std_dta = scaler.transform(input_data_reshaped)
#print(std_dta)

pred = classifier.predict(std_dta)
#print(pred)

if (pred[0]==0):
    print('The person is not diabetic')
else :
    print('The person is diabetic')
```
